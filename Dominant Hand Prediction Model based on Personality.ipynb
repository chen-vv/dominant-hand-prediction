## Data Preprocessing

### Handling Missing Data

Any rows that are missing the target variable will be deleted, since we are dealing with supervised learning and rows with missing target values will not provide any usable information for training our model.

```python
y.info()
```

```python
assert all(X.index == y.index), "Indices of X and y must be the same!"

# Create a combined dataframe to ensure we remove corresponding rows in both X and y
combined_df = X.join(y)

# Drop rows where the target column has missing values
combined_df.dropna(subset=y.columns, inplace=True)

# Split the combined dataframe back to X and y
X = combined_df.drop(columns=y.columns)
y = combined_df[y.columns]

# Check that missing values in y have been handled properly 
y.info()
```

```python
X.info()
```

Note that there are some missing values in both numerical and categorical columns. These missing numerical values will be replaced with the mean, while the missing categorical values will be replaced with the mode.  

```python
# Replace NaN values in numerical columns with the rounded mean of the column,
# rounded to the nearest integer.
for col in X.columns[(X.dtypes == 'float64') | (X.dtypes == 'int64')]:
    mean_val = round(X[col].mean())
    X[col].fillna(mean_val, inplace=True)
    
# Replace NaN values with the most frequently occurring value in that column.
for col in X.columns[(X.dtypes == 'object')]:
    mode_val = X[col].mode()[0]
    X[col].fillna(mode_val, inplace=True)
    
# Check that all missing values have been handled properly
X.info()
```

## Training the Models

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

# importing the models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

We saw previously that the target class is imbalanced with 90% of the entries from right-handed individuals. So for our models we will make sure to pass in parameters that account for this class imbalance and assign larger weights for the 'left handed' class.

```python
models = [
    RandomForestClassifier(class_weight='balanced'),
    LogisticRegression(class_weight='balanced', random_state=123, max_iter=1000),
    KNeighborsClassifier(weights='distance', n_neighbors=3),
    SVC(kernel='rbf', class_weight='balanced'),
    DecisionTreeClassifier(max_depth=10, class_weight='balanced')
]
```

```python
from sklearn.metrics import confusion_matrix

for model in models:
    model.fit(X_train, y_train)
    scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=5)
    model_name = type(model).__name__
    print(f'Averaged prediction accuracy for {model_name} = {np.average(scores)}')
    print(f'Scores: ', scores)

    predictions = model.predict(X_train)
    TN, FP, FN, TP = confusion_matrix(y_train, predictions).ravel()

    recall = TP / (TP + FN)
    print("Recall: %0.4f" % (recall))
    precision = TP / (TP + FP)
    print("Precision: %0.4f" % (precision))
    f1_score = (2 * precision * recall) / (precision + recall)
    print("f1: %0.4f\n" % (f1_score))
```

The best performing models seem to be KNN and RandomForestClassifier. Let's focus on using these two models for our final predictions on the test set.

Since KNN has a parameter `n_neighbors` that can be fine-tuned, we will train different KNN models and compare the results.

```python
n_neighbors = [4, 5, 6, 7, 8]

for n in n_neighbors:
    model = KNeighborsClassifier(weights='distance', n_neighbors=n)
    model.fit(X_train, y_train)
    scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=5)
    print(f'Averaged prediction accuracy for KNN with n_neighbors={n} is: {np.average(scores)}')
    print(f'Scores: ', scores)

    predictions = model.predict(X_train)
    TN, FP, FN, TP = confusion_matrix(y_train, predictions).ravel()

    recall = TP / (TP + FN)
    print("Recall: %0.4f" % (recall))
    precision = TP / (TP + FP)
    print("Precision: %0.4f" % (precision))
    f1_score = (2 * precision * recall) / (precision + recall)
    print("f1: %0.4f\n" % (f1_score))
```

It seems that when we have KNN with n_neighbors >= 5, we get the highest accuracy scores, so that is what we will use in our final model.

Next, we fine-tune our DecisionTreeClassifier.

```python
max_depth = [5, 10, 20, 30, 40, 50]

for n in max_depth:
    model = DecisionTreeClassifier(class_weight='balanced', max_depth=n)
    model.fit(X_train, y_train)
    scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=5)
    print(f'Averaged prediction accuracy for Decision Tree with max_depth={n} is: {np.average(scores)}')
    print(f'Scores: ', scores)

    predictions = model.predict(X_train)
    TN, FP, FN, TP = confusion_matrix(y_train, predictions).ravel()

    recall = TP / (TP + FN)
    print("Recall: %0.4f" % (recall))
    precision = TP / (TP + FP)
    print("Precision: %0.4f" % (precision))
    f1_score = (2 * precision * recall) / (precision + recall)
    print("f1: %0.4f\n" % (f1_score))
```

We see that the highest average accuracy score is when max_depth=30, so we will use that in our final DecisionTreeClassifier model. The scores across each fold of the cross-validation also seem to be consistent with each other when max_depth=30.

## Testing the Final Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay

final_models = [
    KNeighborsClassifier(n_neighbors=5, weights='distance'),
    DecisionTreeClassifier(class_weight='balanced', max_depth=30)
]

for model in final_models:
    model.fit(X_train, y_train)
    X_test_df = pd.DataFrame(X_test, columns=X_train.columns)
    y_pred = model.predict(X_test_df)
    disp = ConfusionMatrixDisplay.from_estimator(
        model,
        X_test_df,
        y_test,
        display_labels=["Left-handed", "Right-handed"],
        values_format="d",
        cmap=plt.cm.Blues,
        colorbar=False,
    );

    model_name = type(model).__name__
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()
    
    print(f'Your Model Accuracy for {model_name} is', accuracy_score(y_test, y_pred)*100, "%")
    predictions = model.predict(X_train)
    TN, FP, FN, TP = confusion_matrix(y_train, predictions).ravel()

    recall = TP / (TP + FN)
    print("Recall: %0.4f" % (recall))
    precision = TP / (TP + FP)
    print("Precision: %0.4f" % (precision))
    f1_score = (2 * precision * recall) / (precision + recall)
    print("f1: %0.4f\n" % (f1_score))
```

## Discussion of Results

In this study, we used a dataset from a 2013 questionnaire to build a machine learning model that could accurately predict whether an individual was right or left-handed based on their hobbies, interests, fears, personalities and opinions. We first preprocessed the dataset by substituting missing values with the mean (for numerical features) or mode (for categorical features) and encoding categorical features to numerical ones. Note that feature scaling was not used because the numerical values only ranged from 1 to 5.

We then split our data into train and test sets using a 80/20 split. Classification models were used to examine which ones to further fine-tune, and KNN and DecisionTreeClassifier was chosen in the end. We optimized n_neighbours for KNN and max_depth for DecisionTreeClassifier, then tested our final optimized models on the test set. A Confusion Matrix was generated for each of those two models.

From our Confusion Matrix shown above we see that our two final models do not do a very good job of detecting left-handed individuals. For example, in our DecisionTreeClassifier Confusion Matrix we see that our model correctly predicted 3 left-handed and 168 right-handed individuals, but made 32 incorrect predictions. 15 of those incorrect predictions were from the model predicting "left handed" when the individual was actually right-handed, and 17 of those incorrect predictions were from the model predicting "right handed" when the individual was actually left-handed.

The original purpose of this study was to determine whether left-handed individuals are more inclined towards certain ways of thinking or have specific interests. Contrary to most stereotypes (e.g. "Left-handed individuals are more artistic and creative" or "Left-handed individuals are smarter"), it seems that there is no clear correlation between an individual's personality and their dominant hand based on the performance of our final two models.

At first, we only selected the KNN model to fine-tune. This proved difficult because the model would simply make all its predictions to be "Right handed". As a result, the DecisionTreeClassifier was also chosen to fine-tune, and we see from above that there is variety in its predictions at the expense of a lower accuracy score. However, because our dataset contained a lot more right-handed individuals than left-handed ones, accuracy might not have been the best metric to evaluate our models - in our case, we also considered metrics such as recall, precision, and f1-score.

In further studies, it would be best to have a larger dataset with more balanced entries between the right-handed and left-handed class. Having a dataset with more entries in total would allow the model to be trained on more data and perhaps perform better on the test set. With a more balanced dataset, the model would have more information about the left-handed class that could also help it make more accurate predictions.
